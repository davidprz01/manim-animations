# EXPRESIÓN ALGEBRAICAS


**SCENA 1:**  Expresiones algebraicas, monomio, trinomio y polinomio y su grado  

```python

from manim import *

TITLE_SIZE = 48
SUBTITLE_SIZE = 36
TERM_SIZE = 44
TITLE_COLOR = BLUE
TERM_COLOR = YELLOW


def make_title(scene, text, pos=UP, size=TITLE_SIZE, color=TITLE_COLOR, run_time=1):
    title = Text(text, font_size=size, color=color).to_edge(pos)
    scene.play(Write(title, run_time=run_time))
    return title


def replace_tex(scene, old, new, run_time=1):
    scene.play(TransformMatchingTex(old, new), run_time=run_time)
    return new


def highlight_terms(tex, color_map):
    tex.set_color_by_tex_to_color_map(color_map)
    return tex
```

```python
%%manim -qm -v Warning Escena1_ExpresionesAlgebraicas

class Escena1_ExpresionesAlgebraicas(Scene):
    def construct(self):

# T?tulo principal
titulo = make_title(self, "Expresiones Algebraicas")
self.wait(1)

        # Definición de expresión algebraica
        expresion_ejemplo = MathTex("3x^2 + 5x - 7", font_size=60, color=YELLOW)
        self.play(FadeIn(expresion_ejemplo))
        self.wait(2)
        
        # Señalar partes de la expresión
        terminos = VGroup(
            MathTex("3x^2", color=GREEN).scale(1.2),
            MathTex("+", color=WHITE).scale(1.2),
            MathTex("5x", color=GREEN).scale(1.2),
            MathTex("-", color=WHITE).scale(1.2),
            MathTex("7", color=GREEN).scale(1.2)
        ).arrange(RIGHT, buff=0.3)
        
        self.play(TransformMatchingTex(expresion_ejemplo, terminos))
        self.wait(2)
        
        # Limpiar para siguiente concepto
        self.play(FadeOut(expresion_ejemplo), FadeOut(terminos))
        self.wait(1)
        
        # MONOMIO
        titulo_monomio = Text("Monomio", font_size=40, color=BLUE)
        titulo_monomio.to_edge(UP, buff=1.5)
        self.play(Transform(titulo, titulo_monomio))
        
        monomio1 = MathTex("5x^3", font_size=55, color=YELLOW)
        monomio2 = MathTex("-2ab^2", font_size=55, color=YELLOW)
        monomio3 = MathTex("7", font_size=55, color=YELLOW)
        
        monomios = VGroup(monomio1, monomio2, monomio3).arrange(DOWN, buff=0.8)
        
        self.play(FadeIn(monomio1))
        self.wait(1.5)
        self.play(FadeIn(monomio2))
        self.wait(1.5)
        self.play(FadeIn(monomio3))
        self.wait(2)
        
        # Destacar: un solo término
        rect_monomio = SurroundingRectangle(monomio1, color=GREEN, buff=0.2)
        self.play(Create(rect_monomio))
        self.wait(1)
        self.play(FadeOut(rect_monomio))
        
        self.play(FadeOut(monomios))
        self.wait(1)
        
        # BINOMIO
        titulo_binomio = Text("Binomio", font_size=40, color=BLUE)
        titulo_binomio.to_edge(UP, buff=1.5)
        self.play(Transform(titulo, titulo_binomio))
        
        binomio1 = MathTex("2x + 5", font_size=55, color=YELLOW)
        binomio2 = MathTex("a^2 - 3b", font_size=55, color=YELLOW)
        
        binomios = VGroup(binomio1, binomio2).arrange(DOWN, buff=0.8)
        
        self.play(FadeIn(binomio1))
        self.wait(1.5)
        self.play(FadeIn(binomio2))
        self.wait(2)
        
        # Destacar: dos términos
        termino1 = MathTex("2x", color=GREEN, font_size=55).move_to(binomio1).shift(LEFT*0.8)
        termino2 = MathTex("5", color=GREEN, font_size=55).move_to(binomio1).shift(RIGHT*0.8)
        
        self.play(
            Indicate(termino1, color=GREEN),
            Indicate(termino2, color=GREEN)
        )
        self.wait(2)
        
        self.play(FadeOut(binomios))
        self.wait(1)
        
        # TRINOMIO
        titulo_trinomio = Text("Trinomio", font_size=40, color=BLUE)
        titulo_trinomio.to_edge(UP, buff=1.5)
        self.play(Transform(titulo, titulo_trinomio))
        
        trinomio1 = MathTex("x^2 + 3x - 4", font_size=55, color=YELLOW)
        trinomio2 = MathTex("2a^2 - 5ab + b^2", font_size=55, color=YELLOW)
        
        trinomios = VGroup(trinomio1, trinomio2).arrange(DOWN, buff=0.8)
        
        self.play(FadeIn(trinomio1))
        self.wait(1.5)
        self.play(FadeIn(trinomio2))
        self.wait(2)
        
        # Destacar: tres términos
        self.play(
            trinomio1.animate.scale(1.2).set_color(GREEN)
        )
        self.wait(1)
        self.play(trinomio1.animate.scale(1/1.2).set_color(YELLOW))
        
        self.play(FadeOut(trinomios))
        self.wait(1)
        
        # POLINOMIO
        titulo_polinomio = Text("Polinomio", font_size=40, color=BLUE)
        titulo_polinomio.to_edge(UP, buff=1.5)
        self.play(Transform(titulo, titulo_polinomio))
        
        polinomio = MathTex("4x^4 - 2x^3 + 5x^2 - x + 8", font_size=48, color=YELLOW)
        self.play(FadeIn(polinomio))
        self.wait(2)
        
        # Mostrar que tiene 4 o más términos
        terminos_poly = VGroup(
            MathTex("4x^4", color=GREEN),
            MathTex("-2x^3", color=GREEN),
            MathTex("+5x^2", color=GREEN),
            MathTex("-x", color=GREEN),
            MathTex("+8", color=GREEN)
        ).arrange(RIGHT, buff=0.3).scale(1.1)
        
        self.play(TransformMatchingTex(polinomio, terminos_poly))
        self.wait(2)
        
        self.play(FadeOut(polinomio))
        self.wait(1)
        
        # GRADO DEL POLINOMIO
        titulo_grado = Text("Grado del Polinomio", font_size=40, color=BLUE)
        titulo_grado.to_edge(UP, buff=1.5)
        self.play(Transform(titulo, titulo_grado))
        
        ejemplo_grado = MathTex("P(x) = 5x^3 + 2x^2 - 4x + 1", font_size=50, color=YELLOW)
        self.play(Write(ejemplo_grado))
        self.wait(2)
        
        # Señalar el exponente más alto
        exponente_mayor = MathTex("3", color=RED, font_size=70)
        exponente_mayor.next_to(ejemplo_grado, UP).shift(LEFT*2.2)
        flecha = Arrow(exponente_mayor.get_bottom(), ejemplo_grado.get_top() + LEFT*2.2, color=RED)
        
        self.play(GrowArrow(flecha), FadeIn(exponente_mayor))
        self.wait(1)
        
        grado_texto = Text("Grado: 3", font_size=40, color=GREEN)
        grado_texto.next_to(ejemplo_grado, DOWN, buff=0.8)
        self.play(Write(grado_texto))
        self.wait(2)
        
        self.play(FadeOut(exponente_mayor), FadeOut(flecha))
        self.wait(1)
        
        # Más ejemplos de grados
        self.play(FadeOut(ejemplo_grado), FadeOut(grado_texto))
        
        ejemplo2 = MathTex("Q(x) = x^5 - 3x^2 + 7", font_size=48, color=YELLOW)
        grado2 = Text("Grado: 5", font_size=36, color=GREEN)
        grupo2 = VGroup(ejemplo2, grado2).arrange(DOWN, buff=0.5)
        
        ejemplo3 = MathTex("R(x) = 2x^2 + 4x - 1", font_size=48, color=YELLOW)
        grado3 = Text("Grado: 2", font_size=36, color=GREEN)
        grupo3 = VGroup(ejemplo3, grado3).arrange(DOWN, buff=0.5)
        
        grupos = VGroup(grupo2, grupo3).arrange(DOWN, buff=1)
        
        self.play(FadeIn(grupo2))
        self.wait(2)
        self.play(FadeIn(grupo3))
        self.wait(2)
        
        # Final
        self.play(FadeOut(grupos), FadeOut(titulo))
        self.wait(1)
```

**ESCENA 2:** Suma de Polinomios

```python
%%manim -qm -v Warning Escena2_SumaPolinomios

class Escena2_SumaPolinomios(Scene):
    def construct(self):

# T?tulo con animaci?n de escritura
titulo = make_title(self, "Suma de Polinomios", run_time=1.5)
self.wait(1)

        # Presentar los dos polinomios con animación de aparición
        p_label = MathTex("P(x) =", font_size=44, color=WHITE)
        p_expr = MathTex("x^3 - 6x^2 + 2x + 4", font_size=44, color=YELLOW)
        polinomio_p = VGroup(p_label, p_expr).arrange(RIGHT, buff=0.3)
        polinomio_p.shift(UP * 1.5)
        
        q_label = MathTex("Q(x) =", font_size=44, color=WHITE)
        q_expr = MathTex("6x^2 + 2x + 4", font_size=44, color=YELLOW)
        polinomio_q = VGroup(q_label, q_expr).arrange(RIGHT, buff=0.3)
        polinomio_q.shift(UP * 0.5)
        
        # Animación: Los polinomios entran desde los lados
        polinomio_p.shift(RIGHT * 10)
        self.play(
            polinomio_p.animate.shift(LEFT * 10),
            run_time=1
        )
        self.wait(1)
        
        polinomio_q.shift(LEFT * 10)
        self.play(
            polinomio_q.animate.shift(RIGHT * 10),
            run_time=1
        )
        self.wait(2)
        
        # Mostrar la operación de suma
        suma_symbol = MathTex("+", font_size=50, color=GREEN)
        suma_symbol.next_to(polinomio_q, LEFT, buff=0.5).shift(LEFT * 0.3)
        self.play(FadeIn(suma_symbol, scale=2))
        self.wait(1)
        
        # Línea divisoria animada
        linea = Line(LEFT * 3.5, RIGHT * 3.5, color=WHITE)
        linea.next_to(polinomio_q, DOWN, buff=0.3)
        self.play(Create(linea))
        self.wait(1.5)
        
        # Transformación: Reorganizar términos por grado
        titulo_paso1 = Text("Alinear términos semejantes", font_size=32, color=ORANGE)
        titulo_paso1.to_edge(UP, buff=0.5)
        self.play(Transform(titulo, titulo_paso1))
        self.wait(1)
        
        # Reorganizar P(x) mostrando cada término
        p_terminos = VGroup(
            MathTex("x^3", color=RED).scale(1.1),
            MathTex("-6x^2", color=BLUE).scale(1.1),
            MathTex("+2x", color=GREEN).scale(1.1),
            MathTex("+4", color=PURPLE).scale(1.1)
        ).arrange(RIGHT, buff=0.5)
        p_terminos.move_to(polinomio_p).shift(RIGHT * 0.5)
        
        self.play(
            ReplacementTransform(p_expr, p_terminos),
            FadeOut(p_label)
        )
        self.wait(1)
        
        # Reorganizar Q(x) alineado con P(x)
        q_terminos = VGroup(
            MathTex("0", color=RED, fill_opacity=0.3).scale(1.1),
            MathTex("+6x^2", color=BLUE).scale(1.1),
            MathTex("+2x", color=GREEN).scale(1.1),
            MathTex("+4", color=PURPLE).scale(1.1)
        ).arrange(RIGHT, buff=0.5)
        q_terminos.move_to(polinomio_q).shift(RIGHT * 0.5)
        
        self.play(
            ReplacementTransform(q_expr, q_terminos),
            FadeOut(q_label),
            FadeOut(suma_symbol)
        )
        self.wait(2)
        
        # Destacar términos semejantes con animaciones de pulso
        for i in range(4):
            self.play(
                Indicate(p_terminos[i], color=WHITE, scale_factor=1.3),
                Indicate(q_terminos[i], color=WHITE, scale_factor=1.3),
                run_time=0.8
            )
            self.wait(0.5)
        
        self.wait(1)
        
        # Proceso de suma término por término
        titulo_paso2 = Text("Sumar términos", font_size=32, color=ORANGE)
        titulo_paso2.to_edge(UP, buff=0.5)
        self.play(Transform(titulo, titulo_paso2))
        self.wait(1)
        
        resultado_terminos = VGroup()
        resultado_y = linea.get_center()[1] - 0.8
        
        # Término x³ (solo hay uno)
        term1_proceso = MathTex("x^3", "+", "0", "=", "x^3", color=RED, font_size=36)
        term1_proceso.move_to([p_terminos[0].get_center()[0], resultado_y - 0.5, 0])
        self.play(Write(term1_proceso))
        self.wait(1)
        
        resultado1 = MathTex("x^3", color=RED).scale(1.1)
        resultado1.move_to([p_terminos[0].get_center()[0], resultado_y, 0])
        self.play(
            TransformFromCopy(term1_proceso[4], resultado1),
            FadeOut(term1_proceso)
        )
        resultado_terminos.add(resultado1)
        self.wait(1)
        
        # Término x² (suma real)
        term2_proceso = MathTex("-6x^2", "+", "6x^2", "=", "0", color=BLUE, font_size=36)
        term2_proceso.move_to([p_terminos[1].get_center()[0], resultado_y - 0.5, 0])
        self.play(Write(term2_proceso))
        self.wait(1.5)
        
        resultado2 = MathTex("0", color=BLUE, fill_opacity=0.3).scale(1.1)
        resultado2.move_to([p_terminos[1].get_center()[0], resultado_y, 0])
        self.play(
            TransformFromCopy(term2_proceso[4], resultado2),
            FadeOut(term2_proceso)
        )
        resultado_terminos.add(resultado2)
        self.wait(1)
        
        # Término x (suma)
        term3_proceso = MathTex("2x", "+", "2x", "=", "4x", color=GREEN, font_size=36)
        term3_proceso.move_to([p_terminos[2].get_center()[0], resultado_y - 0.5, 0])
        self.play(Write(term3_proceso))
        self.wait(1.5)
        
        resultado3 = MathTex("+4x", color=GREEN).scale(1.1)
        resultado3.move_to([p_terminos[2].get_center()[0], resultado_y, 0])
        self.play(
            TransformFromCopy(term3_proceso[4], resultado3),
            FadeOut(term3_proceso)
        )
        resultado_terminos.add(resultado3)
        self.wait(1)
        
        # Término independiente
        term4_proceso = MathTex("4", "+", "4", "=", "8", color=PURPLE, font_size=36)
        term4_proceso.move_to([p_terminos[3].get_center()[0], resultado_y - 0.5, 0])
        self.play(Write(term4_proceso))
        self.wait(1.5)
        
        resultado4 = MathTex("+8", color=PURPLE).scale(1.1)
        resultado4.move_to([p_terminos[3].get_center()[0], resultado_y, 0])
        self.play(
            TransformFromCopy(term4_proceso[4], resultado4),
            FadeOut(term4_proceso)
        )
        resultado_terminos.add(resultado4)
        self.wait(2)
        
        # Limpiar y mostrar resultado final
        self.play(
            FadeOut(p_terminos),
            FadeOut(q_terminos),
            FadeOut(linea),
            FadeOut(resultado2)  # El 0 desaparece
        )
        self.wait(1)
        
        # Reorganizar resultado final
        titulo_final = Text("Resultado Final", font_size=40, color=BLUE)
        titulo_final.to_edge(UP, buff=0.5)
        self.play(Transform(titulo, titulo_final))
        
        # Mover resultado al centro
        resultado_limpio = VGroup(resultado1, resultado3, resultado4)
        self.play(
            resultado_limpio.animate.arrange(RIGHT, buff=0.3).move_to(ORIGIN).scale(1.2)
        )
        self.wait(2)
        
        # Encuadrar el resultado final
        resultado_box = SurroundingRectangle(resultado_limpio, color=GOLD, buff=0.3, corner_radius=0.2)
        self.play(Create(resultado_box))
        self.wait(1)
        
        # Escribir la respuesta completa
        respuesta_final = MathTex("P(x) + Q(x) =", "x^3 + 4x + 8", font_size=50)
        respuesta_final[1].set_color(YELLOW)
        respuesta_final.move_to(ORIGIN)
        
        self.play(
            FadeOut(resultado_limpio),
            FadeOut(resultado_box),
            Write(respuesta_final)
        )
        self.wait(2)
        
        # Efecto final: Destacar resultado con rotación
        self.play(
            respuesta_final[1].animate.scale(1.3).set_color(GOLD),
            rate_func=there_and_back,
            run_time=1.5
        )
        self.wait(1)
        
        # Desvanecimiento final
        self.play(
            FadeOut(respuesta_final),
            FadeOut(titulo)
        )
        self.wait(1)
```

**ESCENA 3:** Resta de Polinomios

```python
%%manim -qm -v Warning Escena3_RestaPolinomios

class Escena3_RestaPolinomios(Scene):
    def construct(self):

# T?tulo con animaci?n
titulo = make_title(self, "Resta de Polinomios", run_time=1.5)
self.wait(1)

        # Presentar los dos polinomios
        p_label = MathTex("P(x) =", font_size=44, color=WHITE)
        p_expr = MathTex("x^3 - 6x^2 + 2x + 4", font_size=44, color=YELLOW)
        polinomio_p = VGroup(p_label, p_expr).arrange(RIGHT, buff=0.3)
        polinomio_p.shift(UP * 1.5)
        
        q_label = MathTex("Q(x) =", font_size=44, color=WHITE)
        q_expr = MathTex("6x^2 + 2x + 4", font_size=44, color=YELLOW)
        polinomio_q = VGroup(q_label, q_expr).arrange(RIGHT, buff=0.3)
        polinomio_q.shift(UP * 0.5)
        
        # Animación: Entrada desde arriba y abajo
        polinomio_p.shift(UP * 5)
        self.play(
            polinomio_p.animate.shift(DOWN * 5),
            run_time=1
        )
        self.wait(1)
        
        polinomio_q.shift(DOWN * 5)
        self.play(
            polinomio_q.animate.shift(UP * 5),
            run_time=1
        )
        self.wait(2)
        
        # Mostrar la operación de resta con efecto especial
        resta_symbol = MathTex("-", font_size=50, color=RED)
        resta_symbol.next_to(polinomio_q, LEFT, buff=0.5).shift(LEFT * 0.3)
        self.play(
            FadeIn(resta_symbol, scale=3),
            resta_symbol.animate.set_color(ORANGE)
        )
        self.wait(1)
        
        # Línea divisoria
        linea = Line(LEFT * 3.5, RIGHT * 3.5, color=WHITE)
        linea.next_to(polinomio_q, DOWN, buff=0.3)
        self.play(Create(linea))
        self.wait(1.5)
        
        # PASO 1: Cambiar signos de Q(x)
        titulo_paso1 = Text("Cambiar signos de Q(x)", font_size=32, color=ORANGE)
        titulo_paso1.to_edge(UP, buff=0.5)
        self.play(Transform(titulo, titulo_paso1))
        self.wait(1)
        
        # Destacar Q(x) para cambiar signos
        rect_q = SurroundingRectangle(q_expr, color=RED, buff=0.15)
        self.play(Create(rect_q))
        self.wait(1)
        
        # Mostrar el cambio de signos con animación
        q_expr_new = MathTex("-6x^2 - 2x - 4", font_size=44, color=ORANGE)
        q_expr_new.move_to(q_expr)
        
        self.play(
            Transform(q_expr, q_expr_new),
            rect_q.animate.set_color(GREEN),
            run_time=1.5
        )
        self.wait(2)
        self.play(FadeOut(rect_q))
        
        # PASO 2: Alinear términos
        titulo_paso2 = Text("Alinear términos semejantes", font_size=32, color=ORANGE)
        titulo_paso2.to_edge(UP, buff=0.5)
        self.play(Transform(titulo, titulo_paso2))
        self.wait(1)
        
        # Reorganizar P(x)
        p_terminos = VGroup(
            MathTex("x^3", color=RED).scale(1.1),
            MathTex("-6x^2", color=BLUE).scale(1.1),
            MathTex("+2x", color=GREEN).scale(1.1),
            MathTex("+4", color=PURPLE).scale(1.1)
        ).arrange(RIGHT, buff=0.5)
        p_terminos.move_to(polinomio_p).shift(RIGHT * 0.5)
        
        self.play(
            ReplacementTransform(p_expr, p_terminos),
            FadeOut(p_label)
        )
        self.wait(1)
        
        # Reorganizar Q(x) con signos cambiados
        q_terminos = VGroup(
            MathTex("0", color=RED, fill_opacity=0.3).scale(1.1),
            MathTex("-6x^2", color=BLUE).scale(1.1),
            MathTex("-2x", color=GREEN).scale(1.1),
            MathTex("-4", color=PURPLE).scale(1.1)
        ).arrange(RIGHT, buff=0.5)
        q_terminos.move_to(polinomio_q).shift(RIGHT * 0.5)
        
        self.play(
            ReplacementTransform(q_expr, q_terminos),
            FadeOut(q_label),
            FadeOut(resta_symbol)
        )
        self.wait(2)
        
        # Destacar términos semejantes con efecto wave
        for i in range(4):
            self.play(
                Wiggle(p_terminos[i], scale_value=1.3, rotation_angle=0.02*TAU),
                Wiggle(q_terminos[i], scale_value=1.3, rotation_angle=0.02*TAU),
                run_time=0.8
            )
            self.wait(0.4)
        
        self.wait(1)
        
        # PASO 3: Sumar términos
        titulo_paso3 = Text("Sumar términos", font_size=32, color=ORANGE)
        titulo_paso3.to_edge(UP, buff=0.5)
        self.play(Transform(titulo, titulo_paso3))
        self.wait(1)
        
        resultado_terminos = VGroup()
        resultado_y = linea.get_center()[1] - 0.8
        
        # Término x^3
        term1_proceso = MathTex("x^3", "+", "0", "=", "x^3", color=RED, font_size=36)
        term1_proceso.move_to([p_terminos[0].get_center()[0], resultado_y - 0.5, 0])
        self.play(Write(term1_proceso))
        self.wait(1)
        
        resultado1 = MathTex("x^3", color=RED).scale(1.1)
        resultado1.move_to([p_terminos[0].get_center()[0], resultado_y, 0])
        self.play(
            TransformFromCopy(term1_proceso[4], resultado1),
            FadeOut(term1_proceso)
        )
        resultado_terminos.add(resultado1)
        self.wait(1)
        
        # Término x² (se cancelan)
        term2_proceso = MathTex("-6x^2", "+", "(-6x^2)", "=", "-12x^2", color=BLUE, font_size=36)
        term2_proceso.move_to([p_terminos[1].get_center()[0], resultado_y - 0.5, 0])
        self.play(Write(term2_proceso))
        self.wait(1.5)
        
        resultado2 = MathTex("-12x^2", color=BLUE).scale(1.1)
        resultado2.move_to([p_terminos[1].get_center()[0], resultado_y, 0])
        self.play(
            TransformFromCopy(term2_proceso[4], resultado2),
            FadeOut(term2_proceso)
        )
        resultado_terminos.add(resultado2)
        self.wait(1)
        
        # Término x
        term3_proceso = MathTex("2x", "+", "(-2x)", "=", "0", color=GREEN, font_size=36)
        term3_proceso.move_to([p_terminos[2].get_center()[0], resultado_y - 0.5, 0])
        self.play(Write(term3_proceso))
        self.wait(1.5)
        
        resultado3 = MathTex("0", color=GREEN, fill_opacity=0.3).scale(1.1)
        resultado3.move_to([p_terminos[2].get_center()[0], resultado_y, 0])
        self.play(
            TransformFromCopy(term3_proceso[4], resultado3),
            FadeOut(term3_proceso)
        )
        resultado_terminos.add(resultado3)
        self.wait(1)
        
        # Término independiente
        term4_proceso = MathTex("4", "+", "(-4)", "=", "0", color=PURPLE, font_size=36)
        term4_proceso.move_to([p_terminos[3].get_center()[0], resultado_y - 0.5, 0])
        self.play(Write(term4_proceso))
        self.wait(1.5)
        
        resultado4 = MathTex("0", color=PURPLE, fill_opacity=0.3).scale(1.1)
        resultado4.move_to([p_terminos[3].get_center()[0], resultado_y, 0])
        self.play(
            TransformFromCopy(term4_proceso[4], resultado4),
            FadeOut(term4_proceso)
        )
        resultado_terminos.add(resultado4)
        self.wait(2)
        
        # Limpiar los ceros
        self.play(
            FadeOut(p_terminos),
            FadeOut(q_terminos),
            FadeOut(linea),
            FadeOut(resultado3),
            FadeOut(resultado4)
        )
        self.wait(1)
        
        # Resultado final
        titulo_final = Text("Resultado Final", font_size=40, color=BLUE)
        titulo_final.to_edge(UP, buff=0.5)
        self.play(Transform(titulo, titulo_final))
        
        # Mover resultado al centro
        resultado_limpio = VGroup(resultado1, resultado2)
        self.play(
            resultado_limpio.animate.arrange(RIGHT, buff=0.3).move_to(ORIGIN).scale(1.3)
        )
        self.wait(2)
        
        # Encuadrar con efecto pulsante
        resultado_box = SurroundingRectangle(resultado_limpio, color=GOLD, buff=0.3, corner_radius=0.2)
        self.play(Create(resultado_box))
        self.play(
            resultado_box.animate.set_color(RED),
            rate_func=there_and_back,
            run_time=1
        )
        self.wait(1)
        
        # Respuesta completa
        respuesta_final = MathTex("P(x) - Q(x) =", "x^3 - 12x^2", font_size=50)
        respuesta_final[1].set_color(YELLOW)
        respuesta_final.move_to(ORIGIN)
        
        self.play(
            FadeOut(resultado_limpio),
            FadeOut(resultado_box),
            Write(respuesta_final)
        )
        self.wait(2)
        
        # Efecto final: Rotación completa
        self.play(
            Rotate(respuesta_final[1], angle=TAU, run_time=2),
            respuesta_final[1].animate.set_color(GOLD)
        )
        self.wait(1)
        
        # Desvanecimiento final
        self.play(
            FadeOut(respuesta_final),
            FadeOut(titulo)
        )
        self.wait(1)
```

**ESCENA 4:** Producto de Polinomios

```python
%%manim -qm -v Warning Escena4_ProductoPolinomios

class Escena4_ProductoPolinomios(Scene):
    def construct(self):

# T?tulo
titulo = make_title(self, "Producto de Polinomios", run_time=1.5)
self.wait(1)

        # Presentar el producto completo
        producto_completo = MathTex(
            "(x^3 - 6x^2 + 2x + 4)", "\\cdot", "(6x^2 + 2x + 4)",
            font_size=44,
            color=YELLOW
        )
        producto_completo.shift(UP * 2)
        
        # Entrada con escala
        self.play(GrowFromCenter(producto_completo))
        self.wait(2)
        
        # PASO 1: Aplicar distributiva
        titulo_paso1 = Text("Aplicar propiedad distributiva", font_size=32, color=ORANGE)
        titulo_paso1.to_edge(UP, buff=0.5)
        self.play(Transform(titulo, titulo_paso1))
        self.wait(1)
        
        # Reducir y mover arriba
        self.play(
            producto_completo.animate.scale(0.7).to_edge(UP, buff=1.2)
        )
        self.wait(1)
        
        # Mostrar las 3 multiplicaciones parciales - BIEN ESPACIADAS
        producto1 = MathTex("6x^5", "-", "36x^4", "+", "12x^3", "+", "24x^2", 
                           font_size=36, color=RED)
        producto1.move_to([0, 1, 0])
        
        producto2 = MathTex("2x^4", "-", "12x^3", "+", "4x^2", "+", "8x", 
                           font_size=36, color=BLUE)
        producto2.next_to(producto1, DOWN, buff=0.5)
        
        producto3 = MathTex("4x^3", "+", "4x^2", "+", "8x", "+", "16", 
                           font_size=36, color=GREEN)
        producto3.next_to(producto2, DOWN, buff=0.5)
        
        # Animaciones de entrada
        self.play(Write(producto1))
        self.wait(1.5)
        self.play(Write(producto2))
        self.wait(1.5)
        self.play(Write(producto3))
        self.wait(2)
        
        # Línea para sumar
        linea = Line(LEFT * 5, RIGHT * 5, color=WHITE)
        linea.next_to(producto3, DOWN, buff=0.4)
        self.play(Create(linea))
        self.wait(1)
        
        # PASO 2: Sumar términos semejantes
        titulo_paso2 = Text("Sumar términos semejantes", font_size=32, color=ORANGE)
        titulo_paso2.to_edge(UP, buff=0.5)
        self.play(Transform(titulo, titulo_paso2))
        self.wait(1)
        
        # Destacar términos del mismo grado con pulsos
        # x^5
        self.play(Indicate(producto1[0], color=YELLOW, scale_factor=1.4))
        self.wait(0.5)
        
        # x^4
        self.play(
            Indicate(producto1[2], color=YELLOW, scale_factor=1.4),
            Indicate(producto2[0], color=YELLOW, scale_factor=1.4)
        )
        self.wait(0.5)
        
        # x^3
        self.play(
            Indicate(producto1[4], color=YELLOW, scale_factor=1.4),
            Indicate(producto2[2], color=YELLOW, scale_factor=1.4),
            Indicate(producto3[0], color=YELLOW, scale_factor=1.4)
        )
        self.wait(0.5)
        
        # x^2
        self.play(
            Indicate(producto1[6], color=YELLOW, scale_factor=1.4),
            Indicate(producto2[4], color=YELLOW, scale_factor=1.4),
            Indicate(producto3[2], color=YELLOW, scale_factor=1.4)
        )
        self.wait(0.5)
        
        # x
        self.play(
            Indicate(producto2[6], color=YELLOW, scale_factor=1.4),
            Indicate(producto3[4], color=YELLOW, scale_factor=1.4)
        )
        self.wait(0.5)
        
        # constante
        self.play(Indicate(producto3[6], color=YELLOW, scale_factor=1.4))
        self.wait(1)
        
        # Resultado final de la suma - BIEN ABAJO
        resultado_suma = MathTex(
            "6x^5", "-", "32x^4", "+", "4x^3", "+", "4x^2", "+", "16x", "+", "16",
            font_size=40,
            color=GOLD
        )
        resultado_suma.next_to(linea, DOWN, buff=0.6)
        
        self.play(Write(resultado_suma, run_time=2.5))
        self.wait(2)
        
        # Encuadrar resultado
        box_resultado = SurroundingRectangle(resultado_suma, color=GOLD, buff=0.3)
        self.play(Create(box_resultado))
        self.wait(1)
        
        # Limpiar todo
        self.play(
            FadeOut(producto_completo),
            FadeOut(producto1),
            FadeOut(producto2),
            FadeOut(producto3),
            FadeOut(linea),
            FadeOut(box_resultado)
        )
        self.wait(1)
        
        # RESULTADO FINAL - CENTRADO
        titulo_final = Text("Resultado Final", font_size=40, color=BLUE)
        titulo_final.to_edge(UP, buff=0.5)
        self.play(Transform(titulo, titulo_final))
        
        # Mover al centro y agrandar
        self.play(
            resultado_suma.animate.move_to(ORIGIN).scale(1.3)
        )
        self.wait(2)
        
        # Marco final brillante
        box_final = SurroundingRectangle(resultado_suma, color=GOLD, buff=0.4, corner_radius=0.2)
        self.play(Create(box_final))
        self.play(
            Flash(resultado_suma, color=GOLD, num_lines=20),
            box_final.animate.set_stroke(width=6)
        )
        self.wait(1)
        
        # Fórmula completa
        respuesta_completa = MathTex(
            "P(x) \\cdot Q(x) =", 
            "6x^5 - 32x^4 + 4x^3 + 4x^2 + 16x + 16",
            font_size=38
        )
        respuesta_completa[1].set_color(YELLOW)
        respuesta_completa.move_to(ORIGIN)
        
        self.play(
            FadeOut(resultado_suma),
            FadeOut(box_final),
            Write(respuesta_completa, run_time=2)
        )
        self.wait(2)
        
        # Efecto final: pulso doble
        for _ in range(2):
            self.play(
                respuesta_completa[1].animate.scale(1.15).set_color(GOLD),
                rate_func=there_and_back,
                run_time=0.7
            )
            self.wait(0.2)
        
        self.wait(1)
        
        # Desvanecimiento
        self.play(
            FadeOut(respuesta_completa),
            FadeOut(titulo)
        )
        self.wait(1)
```

**ESCENA 5:** Producto de dos Binomios

```python
%%manim -qm -v Warning Escena5_ProductoBinomios

class Escena5_ProductoBinomios(Scene):
    def construct(self):

# T?tulo principal
titulo = make_title(self, "Producto de dos Binomios", run_time=1.5)
self.wait(1)

        # FÓRMULA GENERAL
        subtitulo = Text("Fórmula General", font_size=36, color=ORANGE)
        subtitulo.next_to(titulo, DOWN, buff=0.8)
        self.play(FadeIn(subtitulo, shift=DOWN))
        self.wait(1)
        
        # Forma general (x+a)(x+b)
        formula_general = MathTex(
            "(x + a)", "(x + b)",
            font_size=50,
            color=YELLOW
        )
        formula_general.shift(UP * 0.5)
        self.play(Write(formula_general))
        self.wait(2)
        
        # Flecha de transformación
        flecha = Arrow(formula_general.get_bottom(), formula_general.get_bottom() + DOWN * 1.5, 
                      color=GREEN, stroke_width=6)
        self.play(GrowArrow(flecha))
        self.wait(0.5)
        
        # Resultado de la fórmula
        resultado_formula = MathTex(
            "x^2 + (a + b)x + ab",
            font_size=50,
            color=GREEN
        )
        resultado_formula.next_to(flecha, DOWN, buff=0.3)
        self.play(Write(resultado_formula, run_time=2))
        self.wait(2)
        
        # Marco alrededor del resultado
        box_formula = SurroundingRectangle(resultado_formula, color=GREEN, buff=0.3)
        self.play(Create(box_formula))
        self.wait(1.5)
        
        # Limpiar para ejemplos
        self.play(
            FadeOut(subtitulo),
            FadeOut(formula_general),
            FadeOut(flecha),
            FadeOut(resultado_formula),
            FadeOut(box_formula)
        )
        self.wait(1)
        
        # ===== EJEMPLO 1: (x+3)(x+2) =====
        titulo_ejemplo = Text("Ejemplo 1", font_size=36, color=ORANGE)
        titulo_ejemplo.to_edge(UP, buff=1.2)
        self.play(Transform(titulo, titulo_ejemplo))
        self.wait(1)
        
        # Problema
        ejemplo1 = MathTex("(", "x", "+", "3", ")", "(", "x", "+", "2", ")", 
                          font_size=50, color=YELLOW)
        ejemplo1.shift(UP * 1.5)
        self.play(Write(ejemplo1))
        self.wait(2)
        
        # Paso 1: Mostrar las operaciones sin resolver
        paso1_a = MathTex(
            "x^2", "+", "(", "3", "+", "2", ")", "x", "+", "(", "3", "\\cdot", "2", ")",
            font_size=44,
            color=WHITE
        )
        paso1_a.next_to(ejemplo1, DOWN, buff=1)
        
        self.play(TransformFromCopy(ejemplo1, paso1_a, run_time=1.5))
        self.wait(1)
        
        # Destacar la primera operación (3+2)
        rect1 = SurroundingRectangle(VGroup(paso1_a[2:7]), color=RED, buff=0.1)
        self.play(Create(rect1))
        self.play(Indicate(paso1_a[2:7], color=RED, scale_factor=1.3))
        self.wait(1)
        
        # Resolver 3+2 = 5
        paso1_b = MathTex(
            "x^2", "+", "5", "x", "+", "(", "3", "\\cdot", "2", ")",
            font_size=44,
            color=WHITE
        )
        paso1_b.move_to(paso1_a)
        paso1_b[2].set_color(GREEN)  # Resaltar el 5
        
        self.play(
            FadeOut(rect1),
            TransformMatchingTex(paso1_a, paso1_b, key_map={
                "(": "", ")": "", "3": "3", "+": "+", "2": "2"
            }),
            run_time=1.5
        )
        self.wait(1)
        
        # Destacar la segunda operación (3·2)
        rect2 = SurroundingRectangle(VGroup(paso1_b[5:10]), color=RED, buff=0.1)
        self.play(Create(rect2))
        self.play(Indicate(paso1_b[5:10], color=RED, scale_factor=1.3))
        self.wait(1)
        
        # Resolver 3·2 = 6
        paso1_c = MathTex(
            "x^2", "+", "5", "x", "+", "6",
            font_size=44,
            color=WHITE
        )
        paso1_c.move_to(paso1_b)
        paso1_c[2].set_color(GREEN)  # Mantener el 5 verde
        paso1_c[5].set_color(GREEN)  # Resaltar el 6
        
        self.play(
            FadeOut(rect2),
            TransformMatchingTex(paso1_b, paso1_c, key_map={
                "(": "", ")": "", "3": "3", "\\cdot": "", "2": "2"
            }),
            run_time=1.5
        )
        self.wait(1.5)
        
        # Animar todos los términos volviéndolos blancos
        self.play(
            paso1_c[2].animate.set_color(WHITE),
            paso1_c[5].animate.set_color(WHITE),
            run_time=0.5
        )
        self.wait(1)
        
        # Resultado final
        resultado1 = MathTex(
            "(x + 3)(x + 2)", "=", "x^2", "+", "5", "x", "+", "6",
            font_size=48
        )
        resultado1[0].set_color(YELLOW)
        resultado1[2:].set_color(GREEN)
        resultado1.shift(DOWN * 1.5)
        
        self.play(
            FadeOut(paso1_c),
            Write(resultado1, run_time=2)
        )
        self.wait(2)
        
        # Marco final
        box1 = SurroundingRectangle(resultado1, color=GOLD, buff=0.3, corner_radius=0.2)
        self.play(Create(box1))
        self.wait(1.5)
        
        # Limpiar
        self.play(
            FadeOut(ejemplo1),
            FadeOut(resultado1),
            FadeOut(box1)
        )
        self.wait(1)
        
        # ===== EJEMPLO 2: (x-3)(x+2) =====
        titulo_ejemplo2 = Text("Ejemplo 2", font_size=36, color=ORANGE)
        titulo_ejemplo2.to_edge(UP, buff=1.2)
        self.play(Transform(titulo, titulo_ejemplo2))
        self.wait(1)
        
        # Problema con signos negativos
        ejemplo2 = MathTex("(", "x", "-", "3", ")", "(", "x", "+", "2", ")", 
                          font_size=50, color=YELLOW)
        ejemplo2.shift(UP * 1.5)
        self.play(Write(ejemplo2))
        self.wait(2)
        
        # Paso 1: Mostrar operaciones con signos negativos
        paso2_a = MathTex(
            "x^2", "+", "(", "-3", "+", "2", ")", "x", "+", "(", "-3", "\\cdot", "2", ")",
            font_size=44,
            color=WHITE
        )
        paso2_a.next_to(ejemplo2, DOWN, buff=1)
        
        self.play(TransformFromCopy(ejemplo2, paso2_a, run_time=1.5))
        self.wait(1)
        
        # Destacar la suma con negativos (-3+2)
        rect1 = SurroundingRectangle(VGroup(paso2_a[2:7]), color=RED, buff=0.1)
        self.play(Create(rect1))
        self.play(Wiggle(paso2_a[2:7], scale_value=1.3))
        self.wait(1)
        
        # Resolver -3+2 = -1
        paso2_b = MathTex(
            "x^2", "+", "(", "-1", ")", "x", "+", "(", "-3", "\\cdot", "2", ")",
            font_size=44,
            color=WHITE
        )
        paso2_b.move_to(paso2_a)
        paso2_b[3].set_color(YELLOW)  # Resaltar -1
        
        self.play(
            FadeOut(rect1),
            TransformMatchingTex(paso2_a, paso2_b),
            run_time=1.5
        )
        self.wait(1)
        
        # Simplificar (-1)x a -1x
        paso2_c = MathTex(
            "x^2", "-", "1", "x", "+", "(", "-3", "\\cdot", "2", ")",
            font_size=44,
            color=WHITE
        )
        paso2_c.move_to(paso2_b)
        paso2_c[2].set_color(YELLOW)
        
        self.play(
            TransformMatchingTex(paso2_b, paso2_c),
            run_time=1
        )
        self.wait(1)
        
        # Destacar la multiplicación (-3·2)
        rect2 = SurroundingRectangle(VGroup(paso2_c[5:10]), color=RED, buff=0.1)
        self.play(Create(rect2))
        self.play(Wiggle(paso2_c[5:10], scale_value=1.3))
        self.wait(1)
        
        # Resolver -3·2 = -6
        paso2_d = MathTex(
            "x^2", "-", "1", "x", "-", "6",
            font_size=44,
            color=WHITE
        )
        paso2_d.move_to(paso2_c)
        paso2_d[2].set_color(YELLOW)
        paso2_d[5].set_color(YELLOW)
        
        self.play(
            FadeOut(rect2),
            TransformMatchingTex(paso2_c, paso2_d),
            run_time=1.5
        )
        self.wait(1)
        
        # Simplificar 1x a x
        paso2_e = MathTex(
            "x^2", "-", "x", "-", "6",
            font_size=44,
            color=WHITE
        )
        paso2_e.move_to(paso2_d)
        paso2_e[2].set_color(GREEN)
        paso2_e[4].set_color(GREEN)
        
        self.play(
            TransformMatchingTex(paso2_d, paso2_e),
            run_time=1.5
        )
        self.wait(1)
        
        # Volver todo a blanco
        self.play(
            paso2_e[2].animate.set_color(WHITE),
            paso2_e[4].animate.set_color(WHITE),
            run_time=0.5
        )
        self.wait(1)
        
        # Resultado final
        resultado2 = MathTex(
            "(x - 3)(x + 2)", "=", "x^2", "-", "x", "-", "6",
            font_size=48
        )
        resultado2[0].set_color(YELLOW)
        resultado2[2:].set_color(GREEN)
        resultado2.shift(DOWN * 1.5)
        
        self.play(
            FadeOut(paso2_e),
            Write(resultado2, run_time=2)
        )
        self.wait(2)
        
        box2 = SurroundingRectangle(resultado2, color=GOLD, buff=0.3, corner_radius=0.2)
        self.play(Create(box2))
        self.wait(1.5)
        
        self.play(
            FadeOut(ejemplo2),
            FadeOut(resultado2),
            FadeOut(box2)
        )
        self.wait(1)
        
        # ===== EJEMPLO 3: (5x+3)(2x+2) =====
        titulo_ejemplo3 = Text("Ejemplo 3", font_size=36, color=ORANGE)
        titulo_ejemplo3.to_edge(UP, buff=1.2)
        self.play(Transform(titulo, titulo_ejemplo3))
        self.wait(1)
        
        # Problema
        ejemplo3 = MathTex("(", "5x", "+", "3", ")", "(", "2x", "+", "2", ")", 
                          font_size=50, color=YELLOW)
        ejemplo3.shift(UP * 1.5)
        self.play(Write(ejemplo3))
        self.wait(2)
        
        # Paso 1: Aplicar la fórmula mostrando multiplicaciones
        paso3_a = MathTex(
            "(", "5", "\\cdot", "2", ")", "x^2", "+", 
            "(", "(", "5", "\\cdot", "2", ")", "+", "(", "3", "\\cdot", "2", ")", ")", "x", "+",
            "(", "3", "\\cdot", "2", ")",
            font_size=38,
            color=WHITE
        )
        paso3_a.next_to(ejemplo3, DOWN, buff=1)
        
        self.play(TransformFromCopy(ejemplo3, paso3_a, run_time=1.5))
        self.wait(1)
        
        # Destacar y resolver 5·2 del término x²
        rect1 = SurroundingRectangle(VGroup(paso3_a[0:5]), color=RED, buff=0.1)
        self.play(Create(rect1))
        self.play(Indicate(paso3_a[0:5], color=RED, scale_factor=1.2))
        self.wait(1)
        
        # Resolver 5·2 = 10 en x²
        paso3_b = MathTex(
            "10", "x^2", "+", 
            "(", "(", "5", "\\cdot", "2", ")", "+", "(", "3", "\\cdot", "2", ")", ")", "x", "+",
            "(", "3", "\\cdot", "2", ")",
            font_size=38,
            color=WHITE
        )
        paso3_b.move_to(paso3_a)
        paso3_b[0].set_color(YELLOW)
        
        self.play(
            FadeOut(rect1),
            TransformMatchingTex(paso3_a, paso3_b),
            run_time=1.5
        )
        self.wait(1)
        
        # Destacar primera multiplicación en el paréntesis (5·2)
        rect2 = SurroundingRectangle(VGroup(paso3_b[4:9]), color=RED, buff=0.1)
        self.play(Create(rect2))
        self.play(Indicate(paso3_b[4:9], color=RED, scale_factor=1.2))
        self.wait(1)
        
        # Resolver primer 5·2 = 10 en el paréntesis
        paso3_c = MathTex(
            "10", "x^2", "+", 
            "(", "10", "+", "(", "3", "\\cdot", "2", ")", ")", "x", "+",
            "(", "3", "\\cdot", "2", ")",
            font_size=38,
            color=WHITE
        )
        paso3_c.move_to(paso3_b)
        paso3_c[0].set_color(YELLOW)
        paso3_c[4].set_color(YELLOW)
        
        self.play(
            FadeOut(rect2),
            TransformMatchingTex(paso3_b, paso3_c),
            run_time=1.5
        )
        self.wait(1)
        
        # Destacar segunda multiplicación (3·2) dentro del paréntesis
        rect3 = SurroundingRectangle(VGroup(paso3_c[6:11]), color=RED, buff=0.1)
        self.play(Create(rect3))
        self.play(Indicate(paso3_c[6:11], color=RED, scale_factor=1.2))
        self.wait(1)
        
        # Resolver 3·2 = 6 en el paréntesis
        paso3_d = MathTex(
            "10", "x^2", "+", "(", "10", "+", "6", ")", "x", "+",
            "(", "3", "\\cdot", "2", ")",
            font_size=40,
            color=WHITE
        )
        paso3_d.move_to(paso3_c)
        paso3_d[0].set_color(YELLOW)
        paso3_d[4].set_color(YELLOW)
        paso3_d[6].set_color(YELLOW)
        
        self.play(
            FadeOut(rect3),
            TransformMatchingTex(paso3_c, paso3_d),
            run_time=1.5
        )
        self.wait(1)
        
        # Destacar la suma (10+6)
        rect4 = SurroundingRectangle(VGroup(paso3_d[3:8]), color=GREEN, buff=0.1)
        self.play(Create(rect4))
        self.play(Indicate(paso3_d[3:8], color=GREEN, scale_factor=1.2))
        self.wait(1)
        
        # Resolver 10+6 = 16
        paso3_e = MathTex(
            "10", "x^2", "+", "16", "x", "+",
            "(", "3", "\\cdot", "2", ")",
            font_size=40,
            color=WHITE
        )
        paso3_e.move_to(paso3_d)
        paso3_e[0].set_color(YELLOW)
        paso3_e[3].set_color(GREEN)
        
        self.play(
            FadeOut(rect4),
            TransformMatchingTex(paso3_d, paso3_e),
            run_time=1.5
        )
        self.wait(1)
        
        # Destacar última multiplicación (3·2)
        rect5 = SurroundingRectangle(VGroup(paso3_e[6:11]), color=RED, buff=0.1)
        self.play(Create(rect5))
        self.play(Indicate(paso3_e[6:11], color=RED, scale_factor=1.2))
        self.wait(1)
        
        # Resolver última 3·2 = 6
        paso3_f = MathTex(
            "10", "x^2", "+", "16", "x", "+", "6",
            font_size=40,
            color=WHITE
        )
        paso3_f.move_to(paso3_e)
        paso3_f[0].set_color(YELLOW)
        paso3_f[3].set_color(GREEN)
        paso3_f[6].set_color(GREEN)
        
        self.play(
            FadeOut(rect5),
            TransformMatchingTex(paso3_e, paso3_f),
            run_time=1.5
        )
        self.wait(1)
        
        # Volver todo a blanco
        self.play(
            paso3_f[0].animate.set_color(WHITE),
            paso3_f[3].animate.set_color(WHITE),
            paso3_f[6].animate.set_color(WHITE),
            run_time=0.5
        )
        self.wait(1)
        
        # Resultado final con ecuación completa
        resultado3 = MathTex(
            "(5x + 3)(2x + 2)", "=", "10", "x^2", "+", "16", "x", "+", "6",
            font_size=44
        )
        resultado3[0].set_color(YELLOW)
        resultado3[2:].set_color(GREEN)
        resultado3.shift(DOWN * 1.8)
        
        self.play(
            FadeOut(paso3_f),
            Write(resultado3, run_time=2)
        )
        self.wait(2)
        
        box3 = SurroundingRectangle(resultado3, color=GOLD, buff=0.3, corner_radius=0.2)
        self.play(Create(box3))
        self.wait(1.5)
        
        self.play(
            FadeOut(ejemplo3),
            FadeOut(resultado3),
            FadeOut(box3)
        )
        self.wait(1)
        
        # ===== EJEMPLO 4: (5x-3)(2x+2) =====
        titulo_ejemplo4 = Text("Ejemplo 4", font_size=36, color=ORANGE)
        titulo_ejemplo4.to_edge(UP, buff=1.2)
        self.play(Transform(titulo, titulo_ejemplo4))
        self.wait(1)
        
        # Problema con signos negativos
        ejemplo4 = MathTex("(", "5x", "-", "3", ")", "(", "2x", "+", "2", ")", 
                          font_size=50, color=YELLOW)
        ejemplo4.shift(UP * 1.5)
        self.play(Write(ejemplo4))
        self.wait(2)
        
        # Paso 1: Aplicar fórmula con signos negativos
        paso4_a = MathTex(
            "(", "5", "\\cdot", "2", ")", "x^2", "+", 
            "(", "(", "5", "\\cdot", "2", ")", "+", "(", "-3", "\\cdot", "2", ")", ")", "x", "+",
            "(", "-3", "\\cdot", "2", ")",
            font_size=36,
            color=WHITE
        )
        paso4_a.next_to(ejemplo4, DOWN, buff=1)
        
        self.play(TransformFromCopy(ejemplo4, paso4_a, run_time=1.5))
        self.wait(1)
        
        # Destacar y resolver 5·2 del término x²
        rect1 = SurroundingRectangle(VGroup(paso4_a[0:5]), color=RED, buff=0.1)
        self.play(Create(rect1))
        self.play(Indicate(paso4_a[0:5], color=RED, scale_factor=1.2))
        self.wait(1)
        
        # Resolver 5·2 = 10
        paso4_b = MathTex(
            "10", "x^2", "+", 
            "(", "(", "5", "\\cdot", "2", ")", "+", "(", "-3", "\\cdot", "2", ")", ")", "x", "+",
            "(", "-3", "\\cdot", "2", ")",
            font_size=36,
            color=WHITE
        )
        paso4_b.move_to(paso4_a)
        paso4_b[0].set_color(YELLOW)
        
        self.play(
            FadeOut(rect1),
            TransformMatchingTex(paso4_a, paso4_b),
            run_time=1.5
        )
        self.wait(1)
        
        # Destacar primera multiplicación (5·2) en paréntesis
        rect2 = SurroundingRectangle(VGroup(paso4_b[4:9]), color=RED, buff=0.1)
        self.play(Create(rect2))
        self.play(Indicate(paso4_b[4:9], color=RED, scale_factor=1.2))
        self.wait(1)
        
        # Resolver 5·2 = 10
        paso4_c = MathTex(
            "10", "x^2", "+", 
            "(", "10", "+", "(", "-3", "\\cdot", "2", ")", ")", "x", "+",
            "(", "-3", "\\cdot", "2", ")",
            font_size=36,
            color=WHITE
        )
        paso4_c.move_to(paso4_b)
        paso4_c[0].set_color(YELLOW)
        paso4_c[4].set_color(YELLOW)
        
        self.play(
            FadeOut(rect2),
            TransformMatchingTex(paso4_b, paso4_c),
            run_time=1.5
        )
        self.wait(1)
        
        # Destacar multiplicación con negativo (-3·2)
        rect3 = SurroundingRectangle(VGroup(paso4_c[6:11]), color=RED, buff=0.1)
        self.play(Create(rect3))
        self.play(Wiggle(paso4_c[6:11], scale_value=1.3))
        self.wait(1)
        
        # Resolver -3·2 = -6
        paso4_d = MathTex(
            "10", "x^2", "+", "(", "10", "+", "(", "-6", ")", ")", "x", "+",
            "(", "-3", "\\cdot", "2", ")",
            font_size=38,
            color=WHITE
        )
        paso4_d.move_to(paso4_c)
        paso4_d[0].set_color(YELLOW)
        paso4_d[4].set_color(YELLOW)
        paso4_d[7].set_color(RED)
        
        self.play(
            FadeOut(rect3),
            TransformMatchingTex(paso4_c, paso4_d),
            run_time=1.5
        )
        self.wait(1)
        
        # Simplificar paréntesis: +(-6) = -6
        paso4_e = MathTex(
            "10", "x^2", "+", "(", "10", "-", "6", ")", "x", "+",
            "(", "-3", "\\cdot", "2", ")",
            font_size=38,
            color=WHITE
        )
        paso4_e.move_to(paso4_d)
        paso4_e[0].set_color(YELLOW)
        paso4_e[4].set_color(YELLOW)
        paso4_e[6].set_color(RED)
        
        self.play(
            TransformMatchingTex(paso4_d, paso4_e),
            run_time=1
        )
        self.wait(1)
        
        # Destacar la resta (10-6)
        rect4 = SurroundingRectangle(VGroup(paso4_e[3:8]), color=GREEN, buff=0.1)
        self.play(Create(rect4))
        self.play(Indicate(paso4_e[3:8], color=GREEN, scale_factor=1.2))
        self.wait(1)
        
        # Resolver 10-6 = 4
        paso4_f = MathTex(
            "10", "x^2", "+", "4", "x", "+",
            "(", "-3", "\\cdot", "2", ")",
            font_size=38,
            color=WHITE
        )
        paso4_f.move_to(paso4_e)
        paso4_f[0].set_color(YELLOW)
        paso4_f[3].set_color(GREEN)
        
        self.play(
            FadeOut(rect4),
            TransformMatchingTex(paso4_e, paso4_f),
            run_time=1.5
        )
        self.wait(1)
        
        # Destacar última multiplicación (-3·2)
        rect5 = SurroundingRectangle(VGroup(paso4_f[6:11]), color=RED, buff=0.1)
        self.play(Create(rect5))
        self.play(Wiggle(paso4_f[6:11], scale_value=1.3))
        self.wait(1)
        
        # Resolver -3·2 = -6
        paso4_g = MathTex(
            "10", "x^2", "+", "4", "x", "-", "6",
            font_size=40,
            color=WHITE
        )
        paso4_g.move_to(paso4_f)
        paso4_g[0].set_color(YELLOW)
        paso4_g[3].set_color(GREEN)
        paso4_g[6].set_color(RED)
        
        self.play(
            FadeOut(rect5),
            TransformMatchingTex(paso4_f, paso4_g),
            run_time=1.5
        )
        self.wait(1)
        
        # Volver todo a blanco
        self.play(
            paso4_g[0].animate.set_color(WHITE),
            paso4_g[3].animate.set_color(WHITE),
            paso4_g[6].animate.set_color(WHITE),
            run_time=0.5
        )
        self.wait(1)
        
        # Resultado final
        resultado4 = MathTex(
            "(5x - 3)(2x + 2)", "=", "10", "x^2", "+", "4", "x", "-", "6",
            font_size=44
        )
        resultado4[0].set_color(YELLOW)
        resultado4[2:].set_color(GREEN)
        resultado4.shift(DOWN * 1.8)
        
        self.play(
            FadeOut(paso4_g),
            Write(resultado4, run_time=2)
        )
        self.wait(2)
        
        box4 = SurroundingRectangle(resultado4, color=GOLD, buff=0.3, corner_radius=0.2)
        self.play(Create(box4))
        self.play(Flash(resultado4, color=GOLD, num_lines=16))
        self.wait(2)
        
        # Final
        self.play(
            FadeOut(ejemplo4),
            FadeOut(resultado4),
            FadeOut(box4),
            FadeOut(titulo)
        )
        self.wait(1)
```
